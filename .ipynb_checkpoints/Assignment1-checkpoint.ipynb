```python
import matplotlib.pyplot as plt # Show np-array as image
import matplotlib.image as mpimg # Load/read jpg as np-array
import matplotlib.patches as patches # Draw rectagles
import rasterio.features
from shapely.geometry.polygon import Polygon
```

```python
from skimage.transform import resize
def entry_to_arrays(entry, downsize=(256,256)):
    '''Reads an data entry and returns its image_array and mask_array '''
    annotations, image_fn = entry['annotations'], entry['file_name']
    
    # List of all polygons
    Polygon_objects=[]
    for anno in annotations:
        for char in anno:
            polygon = char['polygon']
            Polygon_objects.append(Polygon(polygon))
    
    # Path of jpg files on MLTGPU server: ~/scratch/lt2326-h21/a1/images
    img_file = '/scratch/lt2326-h21/a1/images/'+image_fn
    
    # The features = the image itself
    img_array = mpimg.imread(img_file) #a np-array of (2048,2048,3) ie, (w,h,RGB) 

    # The labels = the mask of whether a pixel is in a box 
    w, h = img_array.shape[:2]
    mask_array = rasterio.features.rasterize(Polygon_objects, out_shape=(w,h))
    
    return resize(img_array, downsize), mask_array

img_array, mask_array = entry_to_arrays(data_entries[49])

print('Features:',img_array.shape, '  Labels:', mask_array.shape)
for img in (img_array, mask_array):
    plt.figure()
    plt.imshow(img)
```

### Models training

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
```

#### train function

```python
def train_model(xy_pairs, batch_size, epochs, device, model, model_fn, lr=0.01):
    '''Trains a model with batched X,Y data'''
    
    # Load list of [(x,y)...] to batches of tensors
    batcher = DataLoader(xy_pairs, shuffle=True, batch_size=batch_size)
    
    # Train a new model or continue training an existing model
    m = model.to(device)
    m.train()
        
    criterion = nn.BCELoss()
    optimizer = optim.Adam(m.parameters(), lr=lr)
    
    # Loop of each epoch
    for e in range(epochs):
        total_loss = 0
        
        # Each batch
        for i, (x,y) in enumerate(batcher):
            optimizer.zero_grad()
            # Put features/labels on GPU
            x, y = x.to(device), y.float().to(device)
            
            outputs = m(x) #model predictions
            
            #Flatten output and labels so they can be compared
            if type(outputs)==tuple: #multiple losses if multiple outputs
                losses = [criterion(out.reshape(-1), y.reshape(-1)) for out in outputs]
                loss = sum(losses)
            else:
                loss = criterion(outputs.reshape(-1), y.reshape(-1))
            
            total_loss += loss.item()
            
            # Delete batch tensors on GPU to release memory
            del x
            del y
            torch.cuda.empty_cache()
            
            # Learn/optimize the model from the loss
            loss.backward()
            optimizer.step()
            
            print(f'Epoch {e+1}  Avg loss: {total_loss/(i+1)}  Processed data: {i+1}', end='\r')
            # Save the trained-so-far model
            torch.save(m.state_dict(), model_fn)
            
        #line break at the end of an epoch    
        print()
    
    return m # returns trained model
```

#### model 1 - simple CNN LeNet
Inspired by the architecture from https://d2l.ai/chapter_convolutional-neural-networks/lenet.html
(but uses max-pooling instead of avg-pooling)

```python
# CNN-LeNet   https://d2l.ai/chapter_convolutional-neural-networks/lenet.html

class ModelCNN(nn.Module):
    def __init__(self, hidden_size=512, image_size=(256,256), hidden_channels=8):
        super(ModelCNN, self).__init__()
        self.hs = hidden_size
        self.hc = hidden_channels 
        self.W, self.H = image_size
        
        # Bx3x2048x2048 => Bx16x1024x1024
        self.layers_1 = nn.Sequential(
            nn.Conv2d(3,self.hc,(3,3), padding=(1,1)), # in_chan, out_chan, 3*3 window, 
            nn.ReLU(), nn.MaxPool2d((4,4)),
        )
        
        # Bx16x1024x1024 => Bx32x512x512
        self.layers_2 = nn.Sequential(
            nn.Conv2d(self.hc,2*self.hc,(3,3), padding=(1,1)),
            nn.ReLU(), nn.MaxPool2d((2,2)),
        )
        
        # Bx32x512x512 => Bx64x128x128
        self.layers_3 = nn.Sequential(
            nn.Conv2d(2*self.hc,4*self.hc,(3,3), stride=1, padding=(1,1)),
            nn.ReLU(), nn.MaxPool2d((2,2)),
            nn.Dropout(0.2)
        )
        
        # Bx64x128x128 => BxHS/2
        self.layers_4 = nn.Sequential(
            nn.Linear(4*self.hc*(self.W//16)*(self.H//16), self.hs), 
            nn.ReLU(), 
            nn.Linear(self.hs, self.hs//2),         
        )
        
        # BxHS/2 => Bx64x2048x2048
        self.layers_5 = nn.Sequential(
            nn.Sigmoid(),
            nn.Upsample((2048,2048)),
        )
        
        
    def forward(self, batch):
        # original tensor: B x W x H x 3
        current_matrix = batch.float().permute(0,3,1,2) # Move channel dim to front =>B x 3 x W x H
        
        #L1:
        current_matrix = self.layers_1(current_matrix) # B x hc x W/2 x H/2
        #L2:
        current_matrix = self.layers_2(current_matrix) # =>B x 2hc x W/4 x H/4
        #L3
        current_matrix = self.layers_3(current_matrix)  # =>B x 4hc x W/16 x H/16
        
        #L4
        B,C,W,H = current_matrix.shape
        current_matrix = current_matrix.reshape(B, C*W*H) # change shape to 1D
        current_matrix = self.layers_4(current_matrix)
        
        #L5
        current_matrix = current_matrix.reshape(B,1,self.hs//2,1) # B x1 x1 xHS/2
        current_matrix = self.layers_5(current_matrix) # =>B x 1 x 2048 x 2048
        
        out = current_matrix.permute(0,2,3,1) # output:  Bx2048x2048x1

        return out
```

#### model 2 - deeper CNN 
With the conv layers to imitate ResNet18 from
https://ravivaishnav20.medium.com/visualizing-feature-maps-using-pytorch-12a48cd1e573 (not the whole architecture, just the 17 conv2d layers)

This model also has 2 outputs for x and y axes 

```python
# https://ravivaishnav20.medium.com/visualizing-feature-maps-using-pytorch-12a48cd1e573

class ModelConv2D(nn.Module):
    def __init__(self, hidden_size=1000):
        super(ModelConv2D, self).__init__()
        self.hs = hidden_size

        self.layers = [] # list of nn functions
        self.layers.extend(self.add_conv_layer(3,64,7,2,3, maxpool_params=(3,2,1,1))) # L1; B,64,128,128
        self.layers.extend(self.add_conv_layer(64,64,3,1,1)*4 )# L2~5; B,64,128,128
        self.layers.extend(self.add_conv_layer(64,128,3,2,1))# L6
        self.layers.extend(self.add_conv_layer(128,128,3,1,1)*3 ) # L7~9
        self.layers.extend(self.add_conv_layer(128,256,3,2,1, drop=0.2))# L10
        self.layers.extend(self.add_conv_layer(256,256,3,1,1)*3 ) # L11~13
        self.layers.extend(self.add_conv_layer(256,512,3,2,1, drop=0.2))# L14
        self.layers.extend(self.add_conv_layer(512,512,3,1,1)*3 )# L15~17
        
        self.layers1_17 = nn.Sequential(*self.layers)
        
        # B,512,1,1
        self.layers_18 = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1))
        )
        
        # B,512 => B,1000
        self.layers_19 = nn.Sequential(
            nn.Linear( 512, self.hs ),
        )
        
        # B,1024 => B,1,1,1024 => B,1,2048,2048
        self.layers_20 = nn.Sequential(
            nn.Sigmoid(),
            nn.Upsample((2048,2048)),
#             nn.Sigmoid(), # maybe apply outside model (on output) to see if there's difference?
        )
        
        
    def add_conv_layer(self, in_c, out_c, kern, stride, padding, maxpool_params=None, drop=None):
        funcs = [nn.Conv2d(in_c,out_c,(kern,kern), stride=stride, padding=padding), 
                nn.BatchNorm2d(out_c), nn.ReLU()]
        if maxpool_params:
            k,s,p,d = maxpool_params
            funcs.append(nn.MaxPool2d((k,k), stride=s, padding=p, dilation=d))
        if drop:
            funcs.append(nn.Dropout(drop))
        return funcs
        
    def forward(self, batch):
        # original tensor: B x W x H x 3
        current_matrix = batch.float().permute(0,3,1,2) # Move channel dim to front =>B x 3 x W x H
        
        #L1~17: layers of (conv2d, batchnorm, relu)
        current_matrix = self.layers1_17(current_matrix) 
        #L18: avgpool
        current_matrix = self.layers_18(current_matrix)
        #L19: flatten to 1D => linear
        B,C,W,H = current_matrix.shape
        current_matrix = current_matrix.reshape(B, C*W*H) # Bx512
        current_matrix = self.layers_19(current_matrix) # =>B x 1000
        #L5: reshape to 2D => sigmoid & upsample
        x_axis = current_matrix.reshape(B, 1, 1, self.hs)
        y_axis = current_matrix.reshape(B, 1, self.hs, 1)
        x_axis, y_axis = self.layers_20(x_axis), self.layers_20(y_axis) # Upsample x y axes separately
        
        return x_axis.permute(0,2,3,1), y_axis.permute(0,2,3,1)
```

```python
#Train model 1
gpu = torch.device('cuda:2')
init_model = ModelCNN(hidden_size=512, image_size=(256,256))
cnn_model = train_model(train_xy, 
                       batch_size=10, epochs=10, device=gpu, 
                       model=init_model, model_fn='cnn_model.pt')
```

```python
#Train model 2
gpu = torch.device('cuda:0')
init_model = ModelConv2D()
conv2d_model = train_model(train_xy, 
                       batch_size=10, epochs=10, device=gpu, 
                       model=init_model, model_fn='conv2d_model.pt')
```

```python
# Evaluate mean squared error
def eval_model(test_xy, model, threshold=0.6):
    
    model.eval()
    batches = DataLoader(test_xy, batch_size=10)
    criterion = nn.MSELoss()
    
    total_loss = 0
    sum_accuracies = 0
    for i, (x,y) in enumerate(batches):
        
        with torch.no_grad():
            outputs = model(x)
            
        if type(outputs)==tuple: #multiple losses if multiple outputs
            preds = [torch.sigmoid(out) for out in outputs]
            losses = [criterion(pred.reshape(-1), y.reshape(-1)) for pred in preds]
            loss = sum(losses)/len(losses)
            heatmap = preds[0]*preds[1]
        else:
            pred = torch.sigmoid(outputs)
            loss = criterion(pred.reshape(-1), y.reshape(-1))
            heatmap = pred

        total_loss += loss.item()
        
        binary = (heatmap > threshold).float().squeeze() # turn pixel values to 1/0
        matches = torch.eq(binary, y).sum() # number of exact 1/0 matches
        accuracy = matches / y.reshape(-1).shape[0] # matches / total number of pixels in batch
        sum_accuracies += accuracy
        
    print(f'Mean squared error: {total_loss}; avg {total_loss/len(batches)}')
    print('Accuracy:', (sum_accuracies/len(batches)).item() ) 
```

```python
eval_model(test_xy, model=cnn_model, threshold=0.50303566)
```
